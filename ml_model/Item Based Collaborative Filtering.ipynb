```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv(r'ratings_data.csv')
df.head(10)
```

```python
for i in df.columns:
    print(df[i].nunique())
```

```python
arr=(df['Product Name'].unique())
for i in arr:
    print(i)
```

```python
rating_counts = pd.DataFrame(df["Product Name"].value_counts())
rating_counts.head()
```

```python
user_product_df = df.pivot_table(index=["User_id"], columns=["Product Name"], values="Rating")

user_product_df.shape
```

```python
user_product_df.head(10)
```

```python
product_name = "Baluchari Saree 2"
product_name = user_product_df[product_name]
```

```python
suggestions=user_product_df.corrwith(product_name).sort_values(ascending=False)
```
