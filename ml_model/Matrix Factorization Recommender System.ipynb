```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"ratings_data.csv")
df.head(10)
```

```python
df=df[['User_id','Product Name','Rating']]
```

```python
df.head(10)
```

```python
product_ratingCount = (df.
     groupby(by = ['Product Name'])['Rating'].
     count().
     reset_index().
     rename(columns = {'Rating': 'totalRatingCount'})
     [['Product Name', 'totalRatingCount']]
    )
product_ratingCount.head(10)
```

```python
rating_with_totalRatingCount = df.merge(product_ratingCount, left_on = 'Product Name', right_on = 'Product Name', how = 'left')
rating_with_totalRatingCount.head(10)
```

```python
user_rating = rating_with_totalRatingCount.drop_duplicates(['User_id','Product Name'])
user_rating.head(10)
```

```python
user_rating.shape
```

```python
product_user_rating_pivot = user_rating.pivot(index = 'User_id', columns = 'Product Name', values = 'Rating').fillna(0)
product_user_rating_pivot.head(10)
```

```python
X = product_user_rating_pivot.values.T
X.shape
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=10, random_state=42)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

```python
product_title = product_user_rating_pivot.columns
product_title_list = list(product_title)
# print(product_title_list)
coffey_hands = product_title_list.index("Teracotta 1")
```

```python
def recom_MatFac():
    corr_coffey_hands  = corr[coffey_hands]
    
    return list(product_title[(corr_coffey_hands >= 0.7)])
```

```python
import pickle

# # Saving model to disk
# pickle.dump(recom_MatFac(), open('MatFac.pkl','wb'))

with open('correlation_matrix.pkl', 'wb') as file:
    pickle.dump(corr, file)
with open('product_title_list', 'wb') as file:    
    pickle.dump(product_title_list, file)
with open('product_title', 'wb') as file:    
    pickle.dump(product_title, file)
```
